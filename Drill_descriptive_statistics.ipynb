```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
bernoulli= np.random.binomial(1, .5, 100)

plt.hist(bernoulli, color='c')

plt.axvline(bernoulli.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(bernoulli.mean() + bernoulli.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(bernoulli.mean() - bernoulli.std(), color='b', linestyle='dashed', linewidth=2) 

plt.show()
```

The standard deviation and the mean are not good indicators of the bernoulli distribution.

```python
binomial = np.random.binomial(20, .7, 100)

plt.hist(binomial, color='c')

plt.axvline(binomial.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(binomial.mean() + binomial.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(binomial.mean() - binomial.std(), color='b', linestyle='dashed', linewidth=2) 

plt.show()
```

You can see the binomal distribution with a success rate of 70% has a negative skew. The mean is not a great indicator of the data, because of the negative skew. 

```python
beta = np.random.beta(1, 10, 100)

plt.hist(beta, color='c')

plt.axvline(beta.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(beta.mean() + beta.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(beta.mean() - beta.std(), color='b', linestyle='dashed', linewidth=2) 

plt.show()
```

In this beta distribution you can see that most of our data lies below our mean. Our standard deviation is not a good indicator of the the variance, because most of our data that lies above the mean is outside the standard deviation, and most of the data that lies below the mean is inside the standard deviation. A good example of a positive skew.

```python
normal = np.random.standard_normal(100)

plt.hist(normal, color='c')

plt.axvline(normal.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(normal.mean() + normal.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(normal.mean() - normal.std(), color='b', linestyle='dashed', linewidth=2) 

plt.show()
```

In this normal distribution, you can see our data is normalized as expected. Equal amounts of data in our standard deviation. The mean and standard deviation accurately desribe the data.

```python
logistic = np.random.logistic(20, 1, 100)

plt.hist(logistic, color='c')

plt.axvline(logistic.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(logistic.mean() + logistic.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(logistic.mean() - logistic.std(), color='b', linestyle='dashed', linewidth=2) 

plt.show()
```

In this logistic distribution, you can see that our data is heavily weighted in the center creating a positive kurtosis. Most of the data lies inside the standard deviation. The mean and the standard deviation describe the data well.

```python
poisson = np.random.poisson(1.0, 100)

plt.hist(poisson, color='c')

plt.axvline(poisson.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(poisson.mean() + poisson.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(poisson.mean() - poisson.std(), color='b', linestyle='dashed', linewidth=2) 

plt.show()
```

The data in our poisson distribution shows a positive skew. Data above the mean does not lie inside the standard deviation, most of it lies outside the standard deviation. While nearly all of the data that lies below our mean is inside the standard deviation. 

```python
var1 = np.random.normal(5, .5, 100)
var2 = np.random.normal(10, 1, 100)
var3 = var1+var2

plt.hist(var3, color='c')

plt.axvline(var3.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(var3.mean() + var3.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(var3.mean() - var3.std(), color='b', linestyle='dashed', linewidth=2) 


plt.show()
```

In this distribution, you can see the data is pretty close to normalized with a small amounts of negative skew and positive kurtosis. Equal amounts of data in our standard deviation. The mean and standard deviation accurately desribe the data.
